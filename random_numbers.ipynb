# Generate Random Latitudes and Logitudes

## The random() Function

```python
ra.random()
# Any randomy number
```

```python
x = 1
# Sets x to 1

latitudes = []
# Creates an empty list 

while x < 11:
    random_lat = ra.randint(-90, 89) + ra.random()
    latitudes.append(random_lat)
    x += 1
# The loop above says that between 1-11 (because x will start at 1, then loop through it in the while loop until the condition is met)
# Because the while loop condition is x<11, it says "while x is less than 11, loop through this"
# The random_lat value is created through calculating a random integer between -90 and 89 and add an other random number of any value 
# Whatever value is made in the random_lat calculation is added to the empty latitudes list after it is created 
# Then the loop says to add one to x before continuing to loop (if we took this out, it would keep going)
# The loop would keep going because x doesn't have a stopping point.  The loop would forever be met because x = 1 and that isn't changing
```

In the code block above, we:

    Assign the variable x to 1.
    Initialize an empty list, latitudes.
    We create a while loop where we generate a random latitude and add it to the list.
    After the random latitude is added to the list we add one to the variable "x".
    The while loop condition is checked again and will continue to run as long as x is less than 11.

## The NumPy and random Modules
One way to generate more than a thousand latitudes and longitudes is to chain the NumPy module to the random module to create an array of latitudes or longitudes between the lowest and highest values, or –90° and 90°, and –180° and 180°, respectively. To accomplish this, we'll use the uniform() function from the random module.|

**NOTE:** The NumPy module has a built-in random module, and supplements the built-in Python random module. There is no need to import the random module if we import the NumPy module, as it's redundant.

```python
# Import the NumPy module.
import numpy as np
```

```python
np.random.uniform(-90.00, 90.00)
# This is the built in random function
# If you want to use the random package, you can shorten it to ra, but you can't in this funciton because 
# the function itself in the package is set to random
```

```python
# Below is the same things as above 
np.random.uniform(low = -90, high = 90)
```

```python
# use teh size parameter to get more than one floating point decimal number between any set values 
np.random.uniform(-90.000, 90.000, size = 50)
# This creates 50 values between the lower and upper bounds set here 

# We can use this for the lat and long generation if we wanted to, but we don't know if it is faster than using a loop 

# I think that the loop will be slower
```

Timing Code

```python
# Import timeit.
import timeit

# The timeit function takes a while to output, be patient! 
```

```python
%timeit np.random.uniform(-90.000, 90.000, size=1500)
```

```python
def latitudes(size):
    latitudes = []
    x = 0
    while x < (size):
        random_lat = ra.randint(-90, 90) + ra.random()
        latitudes.append(random_lat)
        x += 1
    return latitudes
# Call the function with 1500.
%timeit latitudes(1500)

# The loop does take longer
# It may not appear that way in the beginning, because the values at face look the opposite 
# The unit they are measured in matter 
# The mu seconds = microseconds 
# the ms = milisecons 

# microseconds are smaller than miliseconds
```
